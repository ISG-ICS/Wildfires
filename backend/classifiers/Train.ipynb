```
from icrawler.builtin import GoogleImageCrawler
from imageai.Prediction.Custom import ModelTraining
import os,shutil,random
import PIL.Image
import os


def clean_up_data(folder_name):
  files = os.listdir('/content/gdrive/Team Drives/Summer_ML/Dataset/train/'+ folder_name)
  for fileName in files:
    path_to_image = os.path.join('/content/gdrive/Team Drives/Summer_ML/Dataset/train/'+ folder_name, fileName)
    if fileName.endswith('.png'): 
      os.remove(path_to_image)
      continue    
    rgba_image = PIL.Image.open(path_to_image)
    rgb_image = rgba_image.convert('RGB')                  # To RGB
    #rgb_image.save( fileName.split('.')[0] + '.jpg')       # png -> jpg
    
    
def get_image_of(key, folder_name, num=100):
  key = str(key)
  google_crawler = GoogleImageCrawler(storage={'root_dir': '/content/gdrive/Team Drives/Summer_ML/Dataset/train/'+ folder_name})
  google_crawler.crawl(keyword=key, max_num=num)
  
  
def get_test(folder_name):
  if not os.path.exists('/content/gdrive/Team Drives/Summer_ML/Dataset/test/'+ folder_name):
    os.makedirs('/content/gdrive/Team Drives/Summer_ML/Dataset/test/'+ folder_name)
  files = os.listdir('/content/gdrive/Team Drives/Summer_ML/Dataset/train/'+ folder_name)
  for fileName in random.sample(files, int(len(files)*0.2) ):
    path = os.path.join('/content/gdrive/Team Drives/Summer_ML/Dataset/train/'+ folder_name, fileName)
    shutil.move(path, '/content/gdrive/Team Drives/Summer_ML/Dataset/test/'+ folder_name)

def workflow(key, folder_name, num):
  get_image_of(key, folder_name, num)
  clean_up_data(folder_name)
  get_test(folder_name)
```

```
# 8 training classes crawling from google.
workflow('wildfire', 'wildfire', 1000)
workflow('person','person', 1000)
workflow('fireman','fireman', 1000)
workflow('restaurant','restaurant', 1000)
workflow('fire+smoke','smoke', 1000)
workflow('fog','fog',1000)
workflow('sunset','sunset',1000)
workflow('landscape','landscape',1000)

model_trainer = ModelTraining()

model_trainer.setModelTypeAsResNet()

model_trainer.setDataDirectory('/content/gdrive/Team Drives/Summer_ML/Dataset')

model_trainer.trainModel(num_objects=8, num_experiments=100, enhance_data=True, batch_size=32, show_network_summary=True, transfer_from_model="/content/gdrive/Team Drives/Summer_ML/Combine/resnet50_weights_tf_dim_ordering_tf_kernels.h5",initial_num_objects=1000)

```

```
# Get data
get_image_of('wildfire', 'wildfire', 1000)
get_image_of('fire', 'fire', 1000)
get_image_of('fire+smoke','smoke', 1000)
get_image_of('fireman+wildfire','fireman',1000)
```

```
# Refine data
clean_up_data('fire')
clean_up_data('fireman')
clean_up_data('smoke')
clean_up_data('wildfire')
```

```
get_test('fire')
get_test('fireman')
get_test('wildfire')
get_test('smoke')
```
